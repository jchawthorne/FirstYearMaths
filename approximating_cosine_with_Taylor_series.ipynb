Great, we have our tools.  Now we can think about what we actually want to do.  We have a function

$$f(x) = \cos(x).$$

We want to approximate that function in the region of $x=\pi$

We have seen that we can approximate a function $f(x)$ about $x=a$ as
$$f(x) = f(a) + (x-a) \frac{df}{dx}(a) + \frac{(x-a)^2}{2!}\frac{d^2f}{dx^2}(a)
+\frac{(x-a)^3}{3!}\frac{d^3f}{dx^3}(a) + 
\frac{(x-a)^4}{4!}\frac{d^4f}{dx^4}(a) + \frac{(x-a)^5}{5!}\frac{d^5f}{dx^5}(a) +... $$

<br><br>

Let's begin with the first-order approximation of $f(x)$ near $x=\pi$:

$$f(x) \approx f(a) + (x-a) \frac{df}{dx}(a)$$

We may note that
$$f(\pi) = \cos(\pi) = -1$$
$$\frac{df}{dx}(x=\pi) = -sin(x=\pi) = 0$$

<br>
Our approximation then gives 
$$f(x) \approx -1 + (x-a) \times 0 = -1$$

Let's compute and plot that approximation.  Place your cursor in the cell below and click run.  Then place your cursor in the second cell below and click run.  And finally, place your cursor in the third cell below and click run.

Also make sure you understand what the lines in the cell below are doing.  (You don't need to understand allthe plotting lines.)

```python
# first, we'll make a list of potential x values
# this creates a list of values between -pi and 3*pi, 
# including [-pi/2,-pi/2+pi/1000,-pi/2+2*pi/1000,-pi/2+3*pi/1000,...]
x=arange(-pi/2,3*pi/2,pi/1000)

# we'll also note the 

print('The first 10 values of x to consider are')
print(x[0:10])

# next, let's calculate the true values of cos(x)
f_true = cos(x)

print('The first 10 true values of f(x) are')
print(f_true[0:10])

# note that here f_true is a list of values, 
# the first value in the list f_true is cos(the first value in the list x)
# f_true[0]=cos(x[0])
# the second value in the list f_true is cos(the second value in the list x)
# f_true[1]=cos(x[1])
# and so on

# finally, let's calculate our first-order approximation
f_app1 = -1 + x*0

print('The first 10 approximated values of f(x) are, using a linear approximation')
print(f_app1[0:10])
```

```python
# For now, we want to focus on a small range  of x values, between pi/2 and 3*pi/2
# Let's note this so we only plot a small range
xlimit=[pi/2,3*pi/2]
```

```python
# now let's plot these values

# you need to understand the next two lines of code

# plot the original f(x)
plot(x,f_true,label='true f(x)')

# plot the approximated f(x)
plot(x,f_app1,label='first order approximation')

# but everything past here is just for plotting
# you don't need to understand it
plt.xlim(xlimit);
plt.ylim([-1.5,1.5]);
plt.xticks([pi/2,.75*pi,pi,1.25*pi,1.5*pi])
plt.plot([pi,pi],[-10,10],linestyle=':',color='k',zorder=0)
plt.xlabel('x');
plt.ylabel('f(x)');
plt.legend();
```

That approximation doesn't seem great.  Maybe we should add a second-order component to our approximation.  We now have

$$f(x) \approx f(a) + (x-a) \frac{df}{dx}(a) + \frac{(x-a)^2}{2!}\frac{d^2f}{dx^2}(a)$$

and we may note that
$$f(\pi) = \cos(\pi) = -1$$
$$\frac{df}{dx}(x=\pi) = -\sin(x=\pi) = 0$$
$$\frac{d^f}{dx^2}(x=\pi) = -\cos(x=\pi) = 1$$

So our approximation is 
$$f(x) \approx -1 + (x-a) \times 0 + \frac{(x-a)^2}{2!} \times 1$$

$$f(x) \approx -1 +  \frac{(x-a)^2}{2!} $$


Go ahead and calculate this approximation.  Run the cell below to do the calculation, and make sure you understand it.  Then run the cell afterward to plot the approximation.

Note that to calculate $(x-\pi)^2$, you type power(x-pi,2).

```python
# calculate the approximation value for each value of x
# again, x is a list of values for x
# f_app2 is a list of values for f, with one value corresponding to each value of x
f_app2 = -1 + power(x-pi,2)/2

print('The first 10 values of x are')
print(x[0:10])

print('The first 10 values of the second-order approximation are')
print(f_app2[0:10])
```

```python
# now let's plot these values

# plot the original f(x)
plot(x,f_true,label='true f(x)')

# plot the approximated f(x) with a first-order approximation
plot(x,f_app1,label='first order approximation')

# plot the approximated f(x) with a second-order approximation
plot(x,f_app2,label='second order approximation')

# but everything past here is just for plotting
# you don't need to understand it
plt.xlim(xlimit);
plt.ylim([-1.5,1.5]);
plt.xticks([pi/2,.75*pi,pi,1.25*pi,1.5*pi])
plt.plot([pi,pi],[-10,10],linestyle=':',color='k',zorder=0)
plt.xlabel('x');
plt.ylabel('f(x)');
plt.legend();
```

That second-order approximation seems quite a lot better, particularly at values close to $\pi$.
